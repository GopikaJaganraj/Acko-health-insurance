**Data Columns Information:**  
**1. id (Customer ID):** A unique identifier assigned to each customer in the dataset.
    
**2. Age (Customer Age):** The age of the customer in years at the time of policy purchase.
    
**3. Gender (Customer Gender):** The gender identity of the customer, which can be "Man" or "Woman."
    

**4. Annual Income (Yearly Earnings in INR):** The total income earned by the customer in a year, measured in Indian Rupees (INR).
    

**5. Marital Status (Customer’s Marital Condition):** The marital status of the customer, such as "Spouse Present," "Not Married," or "Formerly Married."
    

**6. Number of Dependents (People Financially Dependent on Customer):** The number of dependents (such as children, parents, or others) that rely on the customer financially.
    

**7. Education Level (Highest Education Attained):** The highest level of education completed by the customer, such as "Undergraduate" or "Post Graduate."
    
**8. Occupation (Customer's Job Type):**The profession or employment category of the customer. In some cases, this data may be missing.
    

**9. Health Score (Overall Health Indicator):** A numerical score representing the customer’s health condition based on lifestyle factors and medical history.
    

**10. Location (Customer’s Residence Tier):** The classification of the customer's residence area into tiers such as Tier-1, Tier-2, or Tier-3 cities.
    
**11. Policy Type (Type of Insurance Policy Chosen):** The category of insurance policy purchased by the customer, such as "Basic," "Premium," or "Comprehensive."
    

**12. Previous Claims (Number of Past Insurance Claims):** The total number of insurance claims the customer has made before purchasing the current policy.
    

**13. Credit Score (Financial Responsibility Indicator):** A numerical representation of the customer’s creditworthiness, indicating their ability to manage finances and make timely payments.
    
**14. Insurance Duration (Policy Tenure in Years):** The number of years the customer has held the insurance policy.
    

**15. Policy Start Date (Date When Policy Became Active):** The date when the insurance policy was purchased and became active.
    

**16. Customer Feedback (Customer Satisfaction Rating):** The rating or feedback provided by the customer about their experience with the insurance policy.
   

**17. Smoking Status (Whether the Customer Smokes):** Indicates whether the customer is a smoker or not, as smoking impacts health risks and insurance premiums.
    

**18. Exercise Frequency (How Often the Customer Exercises):** The frequency of physical exercise performed by the customer, which can impact their health score.
    
**19. Property Type (Type of Residence):** The type of home the customer resides in, such as a detached home or an apartment.
    

**20. Premium Amount (Final Insurance Premium in INR):** The amount the customer pays for their health insurance policy, measured in Indian Rupees (INR).

**5. Creating New Columns in the dataset.**

```python
# Step 1: Create Risk Level Based on Health Score, Smoking, and Exercise
def calculate_risk_level(row):
    if row["Health Score"] < 30 or row["Smoking Status"] == "Yes" or row["Exercise Frequency"] == "Rarely":
        return "High Risk"
    elif row["Health Score"] < 50:
        return "Medium Risk"
    else:
        return "Low Risk"

data["Risk Level"] = data.apply(calculate_risk_level, axis=1)

# Step 2: Calculate Premium to Income Ratio
data["Premium to Income Ratio"] = data["Premium Amount"] / data["Annual Income"]

# Step 3: Create Long-Term Customer Column (1 if more than 5 years, else 0)
data["Long-Term Customer"] = (data["Insurance Duration"] > 5).astype(int)

# Verify new columns
print(data[["Risk Level", "Premium to Income Ratio", "Long-Term Customer"]].head())
```

# **3. Data Cleaning and Preparation**

**1. Converting the datatype for 'Policy Start Date' from object to datetime format.**

```python
data['Policy Start Date'] = pd.to_datetime(data['Policy Start Date'], errors='coerce')  #covert the datatype of the column. Here errors='coerce' this is used to convert the invalid values to nan values without raising any erroe.
data.dtypes  #Checking for the datatypes.
```

**2. Filling the missing values in the numerical columns with the median value of the columns.**

```python
numeric_columns=['Age','Annual Income','Number of Dependents','Health Score','Previous Claims','Credit Score','Premium Amount','Insurance Duration','Premium to Income Ratio']  #array of numerical values.
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].median()) # filling missing values with median
```

**3. Filling the missing values of the Categorical columns with the most frquent values of the current columns.**

```python
columns_to_impute = ['Customer Feedback']  # specifying columns we want to process
for column in columns_to_impute: # Iterate through each column
    most_frequent_value = data[column].mode()[0]   # Find the most frequent value for the current column
    data[column] = data[column].fillna(most_frequent_value)  # Fill missing values in the current column with the most frequent value
```

```python
data.isna().sum()
#Final Check before proceeding to the next step of the analysis, if the missing values are removed or not.
```

**4. Detecing for the Outliers of the numerical columns.**

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # 25th percentile
    Q3 = data[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR  # Lower threshold
    upper_bound = Q3 + 1.5 * IQR  # Upper threshold

    # Return only outlier rows
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers
# Identify numerical columns
numeric_columns = data.select_dtypes(include=['number']).columns  # Only numerical data

# Dictionary to store count of outliers per column
outlier_counts = {}

# Check for outliers in each numerical column
for col in numeric_columns:
    outliers = detect_outliers_iqr(data, col)
    outlier_counts[col] = len(outliers)

# Print the summary of outliers
print("\nTotal Outliers in Each Column:")
print(outlier_counts)
```

**5. Handling the Outliers by capping the values.**

```python
outlier_columns = ["Annual Income", "Previous Claims", "Premium Amount",'Premium to Income Ratio']
# Step 1: Cap extreme values (Winsorization) - only for detected outlier columns
for col in outlier_columns:
    lower_limit = data[col].quantile(0.05)  # 5th percentile
    upper_limit = data[col].quantile(0.95)  # 95th percentile
    data[col] = data[col].clip(lower=lower_limit, upper=upper_limit)  # Capping

print("Extreme Outliers Capped (Winsorization Applied)")

# Step 2: Apply log transformation to highly skewed numerical columns
skewed_columns = ["Annual Income", "Premium Amount"]  # Skewed data needs log transformation

for col in skewed_columns:
    data[col] = np.log1p(data[col])  # log1p prevents log(0) issues

print("Log Transformation Applied to Reduce Outlier Effect")
```

*  Customers with previous claims history tend to be charged higher premiums.

**3. Impact of Smoking & Exercise on Premiums**

*    Smokers tend to have higher premiums compared to non-smokers due to their increased health risks.
*  Customers who exercise more frequently generally have lower premiums, indicating insurers consider lifestyle habits in pricing.

**4. Credit Score & Premium Pricing**
*   Individuals with higher credit scores tend to have lower premium
amounts, suggesting insurers view them as financially responsible and lower risk.
*   There is a strong correlation between credit score and affordability, indicating credit risk is a key factor in premium pricing.


**5. Annual Income vs. Premium Affordability**

*    High-income individuals generally have higher premiums, but not always in proportion to their earnings.
*    Some low-income individuals have unexpectedly high premiums, possibly due to their health risk profiles.
